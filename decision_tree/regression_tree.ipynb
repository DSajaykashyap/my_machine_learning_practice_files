```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Movie_regression.csv", header=0)
```

```python
df.head()
```

```python
df.info()
```

### Missing Value Imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

### Dummy Variable Creation

```python
df.head()
```

```python
df = pd.get_dummies(df,columns = ["3D_available","Genre"],drop_first = True)
```

```python
df.head()
```

### X-y split

```python
X = df.loc[:,df.columns!="Collection"]
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
y = df["Collection"]
type(y)
```

```python
y.head()
```

```python
y.shape
```

### Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

### Traing Regression Tree

```python
from sklearn import tree
regtree = tree.DecisionTreeRegressor(max_depth = 3)
```

```python
regtree.fit(X_train, y_train)
```

### Predict values using trained model

```python
y_train_pred = regtree.predict(X_train)
y_test_pred = regtree.predict(X_test)
```

```python
y_test_pred
```

### Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

### Plotting decision tree

```python
X_train.columns
```

```python
plt.figure(figsize=(25,10))
a=tree.plot_tree(regtree, 
              feature_names= X_train.columns, 
              filled=True, 
              rounded=True, 
              fontsize=14)
```

## Controlling Tree growth

### Maximum number of levels in tree

```python
regtree1 = tree.DecisionTreeRegressor(max_depth = 3)
regtree1.fit(X_train, y_train)
plt.figure(figsize=(25,10))
a=tree.plot_tree(regtree1, 
              feature_names= X_train.columns, 
              filled=True, 
              rounded=True, 
              fontsize=14)
```

### Minimum observations at internal node

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_split = 40)
regtree2.fit(X_train, y_train)
plt.figure(figsize=(25,10))
a=tree.plot_tree(regtree2, 
              feature_names= X_train.columns, 
              filled=True, 
              rounded=True, 
              fontsize=14)
```

### Minimum observations at leaf node

```python
regtree3 = tree.DecisionTreeRegressor(min_samples_leaf = 25, max_depth=4)
regtree3.fit(X_train, y_train)
plt.figure(figsize=(25,10))
a=tree.plot_tree(regtree3, 
              feature_names= X_train.columns, 
              filled=True, 
              rounded=True, 
              fontsize=14)
```
